```python
import numpy as np
import pandas as pd
# Read acsv file
# Check the first five rows of dataframe

data = pd.read_csv('data/data.csv', index_col=5, parse_dates=['Date'])
data.head()
```

```python
data_global = data[data['Region'] == 'global']
groupby_tracks =  data_global.groupby(['Date','Track Name','Artist']).sum()[['Streams']].reset_index()
top10_global = groupby_tracks.sort_values(by=['Streams'], ascending=False)
top_global = top10_global.drop_duplicates(['Track Name']).reset_index(drop=True)
```

```python
# Patch the top50 global dataset

top50_global = groupby_tracks.sort_values(by=['Streams'], ascending=False)
top50_global = top50_global.drop_duplicates(['Track Name']).reset_index(drop=True).head(50)
top50_global.head()
```

```python
# Count amount of the top50 tracks

all_top50_artists = top_global.loc[top_global['Artist'].isin(top50_global['Artist'])]
num_50tracks=  all_top50_artists.groupby(['Artist']).count()[['Track Name']]
total_50tracks_count = num_50tracks.rename(columns={'Track Name':'Total Tracks'})
top50_tracks_count = top50_global.groupby(['Artist']).count()[['Track Name']]
top50_tracks_count = top50_tracks_count.rename(columns={'Track Name':'top50_tracks_count'})
count_50table= top50_tracks_count.merge(total_50tracks_count,on=('Artist'))

count_50table.head()
```

```python
# Add the total of streams in top50

top50_total_streams = top50_global.groupby(['Artist']).sum()[['Streams']]
modify_top50_total_streams = top50_total_streams.rename(columns={'Streams':'top50_total_streams'})
count_50table = count_50table.merge(modify_top50_total_streams,on=('Artist'))
count_50table
```

```python
# def Z_ScoreNormalization(x):
# for x in top50_total_streams:
#     x = float(x - top50_total_streams.mean())/top50_total_streams.std()
#     print(x)
top50_total_streams_norm = (top50_total_streams - top50_total_streams.min()) / (top50_total_streams.max() - top50_total_streams.min())
top50_total_streams_norm
```

```python
# Add the total of streams in top50

# top50_total_streams = top50_global.groupby(['Artist']).sum()[['Streams']]
modify_top50_total_streams = top50_total_streams_norm.rename(columns={'Streams':'top50_total_streams'})
count_50table = count_50table.merge(modify_top50_total_streams,on=('Artist'))
count_50table
```

```python
# install sklearn and statsmodels packages
from sklearn import linear_model
import tkinter as tk
import statsmodels.api as sm
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt

X = count_50table[['Total Tracks', 'top50_total_streams']]
Y = count_50table['top50_tracks_count']

# use sklearn to make regression
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print ('Intercept: \n', regr.intercept_)
print ('Coefficients: \n', regr.coef_)

# by using statsmodels
# add a constant value X
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)

# add tkinter GUI for the task
root = tk.Tk()
root.title('Multiple Linear Regression')

canvas1 = tk.Canvas(root, width=1300, height=420)
canvas1.pack()

intercept_result = ('Intercept: ', regr.intercept_)
label_intercept = tk.Label(root, text=intercept_result, justify = 'center')
canvas1.create_window(260, 270, window=label_intercept)

coefficients_result = ('Coefficients: ', regr.coef_)
label_coefficients = tk.Label(root, text=coefficients_result, justify = 'center')
canvas1.create_window(260, 290, window=label_coefficients)

# add model salary to roght side of the UI
print_model = model.summary()
label_model = tk.Label(root, text=print_model, justify = 'center')
canvas1.create_window(800, 270, window=label_model)

# add totoal tracks and top50_total_streams labels and input boxes
label_tracks = tk.Label(root, text='Input Total Tracks: ')
canvas1.create_window(100, 150, window=label_tracks)
# create tracks entry box
entry_tracks = tk.Entry(root)
canvas1.create_window(350, 150, window=entry_tracks)

label_streams = tk.Label(root, text='Input Top50 Total Streams(millions): ')
canvas1.create_window(140, 200, window=label_streams)
# create tracks entry box
entry_streams = tk.Entry(root)
canvas1.create_window(350, 200, window=entry_streams)

def values(): 
    global Total_Tracks 
    #our 1st input variable
    Total_Tracks = float(entry_tracks.get()) 
    
    global Top50_Total_Streams 
    #our 2nd input variable
    Top50_Total_Streams = float(entry_streams.get()) 
    
    Prediction_result  = ('Predicted Top50 Tracks Count: ', regr.predict([[Total_Tracks ,Top50_Total_Streams]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 360, window=label_Prediction)
    
button_submit = tk.Button (root, text='Predicted Top50 Tracks Count',command=values, bg='orange') 
# button to call the 'values' command above 
canvas1.create_window(270, 240, window=button_submit)

#plot 1st scatter 
figure_tracks = plt.Figure(figsize=(5,4), dpi=100)
ax_tracks = figure_tracks.add_subplot(111)
ax_tracks.scatter(count_50table['Total Tracks'].astype(float),count_50table['top50_tracks_count'].astype(float), color = 'r')
scatter_tracks = FigureCanvasTkAgg(figure_tracks, root) 
scatter_tracks.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax_tracks.legend() 
ax_tracks.set_xlabel('Total Tracks')
ax_tracks.set_title('Total Tracks Vs. Top50 Tracks Count')

#plot 2nd scatter 
figure_streams = plt.Figure(figsize=(5,4), dpi=100)
ax_streams = figure_streams.add_subplot(111)
ax_streams.scatter(count_50table['top50_total_streams'].astype(float),count_50table['top50_tracks_count'].astype(float), color = 'g')
scatter_streams = FigureCanvasTkAgg(figure_streams, root) 
scatter_streams.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax_streams.legend() 
ax_streams.set_xlabel('Top50 Total Streams')
ax_streams.set_title('Top50 Total Streams Vs. Top50 Tracks Count')
 

root.mainloop()
# need to pop the window to view the GUI frame
```
