```python
# hkust vislab code challenge
# This one is to implement a Year/Month HeatMap and daily temperature changes
# Level 2 Challenge
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mplcursors as mp
from matplotlib.widgets import Button
from matplotlib import colors

# first step to read all data from csv file
df = pd.read_csv('temperature_daily.csv')

# add two columns of data field 'year' and 'year-month'
df['year'] = pd.DatetimeIndex(df['date']).year
df['year_month'] = pd.to_datetime(df['date']).dt.to_period('M')
df['month'] = pd.DatetimeIndex(df['date']).month
df['day'] = pd.to_datetime(df['date']).dt.day
df1 = df.set_index(['year'])
df2 = df1.loc['2008':'2018',['year_month','max_temperature','min_temperature', 'month', 'day']]


# generate two pivot table to store the dict of data
table_min = pd.pivot_table(data=df2, index='month', columns='year', values='min_temperature', aggfunc=np.min)
table_max = pd.pivot_table(data=df2, index='month', columns='year', values='max_temperature', aggfunc=np.max)
table_date_min = df2.pivot(index='year_month', columns='day', values='min_temperature')
table_date_max = df2.pivot(index='year_month', columns='day', values='max_temperature')

index = ['January', 'Febuary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
cols = table_min.columns

# solve the table date minimum value and max value
def data_transfer(data_min, data_max):
    min_value = []
    max_value = []
    for col in range(len(table_min.columns)):
        min_value.append([])
        max_value.append([])
        for row in range(len(table_min.index)):
            if row+col*len(table_min.index) < len(data_min):
                min_value[col].append(data_min[row+col*len(table_min.index)])
                max_value[col].append(data_max[row+col*len(table_min.index)])
            else:
                break
    return min_value, max_value

def getHoverText():
    hovertext = list()
    for yi, yy in enumerate(table_min.index):
        hovertext.append(list())
        for xi, xx in enumerate(table_min.columns):
            value_min = table_min.values[yi][xi].astype(int)
            value_max = table_max.values[yi][xi].astype(int)
            hovertext[-1].append('Year: {}\n Month: {}\n MinValue: {}\n MaxValue: {}'.format(xx, yy, value_min, value_max))
    return hovertext

# define subplots
fig, ax = plt.subplots(figsize=(8, 8))
min_date_value, max_date_value = data_transfer(table_date_min.values, table_date_max.values)

# define heatmap
cmap = colors.ListedColormap(["#5E4FA2","#3288BD","#66C2A5","#ABDDA4","#E6F598","#FFFFBF","#FEE08B","#FDAE61","#F46D43","#D53E4F","#9E0142"])
bounds = [0, 5, 10, 15, 20, 25, 30, 35, 40]
norm = colors.BoundaryNorm(bounds, cmap.N)

def heatmap (data, row_labels, col_labels, ax=None, **kwargs):
    if not ax:
        ax = plt.gca()

    # draw heatmap
    # make a color map of fixed colors
    im = ax.imshow(data, **kwargs)
    fig.colorbar(im,cmap=cmap, norm=norm, boundaries=bounds)

    # set the xticks and yticks of heatmap
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)
    #ax.set_ylabel('Year')
    #ax.set_xlabel('Month')

    # Turn spines off and create white grid.
    for edge, spine in ax.spines.items():
        spine.set_visible(False)

    # horizontal axes labeling appear on top
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)

    # set linewidth of heatmap
    ax.set_xticks(np.arange(data.shape[1] + 1) - .5, minor=True)
    ax.set_yticks(np.arange(data.shape[0] + 1) - .5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)

    return im

# define subplot linecharts
def sublinechart(im, data=None, **kwargs):

    # pass the data to linechart
    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # define each linchart height and width
    height = 1.0 / float(data.shape[0])
    width = 1.0 / float(data.shape[1])

    # draw the first graph
    axin = ax.inset_axes([0, 1 - height, width, height], facecolor='None')
    # hide xaxis and yaxis
    axin.get_yaxis().set_visible(False)
    axin.get_xaxis().set_visible(False)

    for edge, spine in axin.spines.items():
        spine.set_visible(False)

    axin.plot(min_date_value[0][0], color='blue', linewidth=1)
    # maximum value
    axin.plot(max_date_value[0][0], color='green', linewidth=1)


    # loop to create each ceil with a line chart
    for col in range(data.shape[1]):
        for row in range(data.shape[0]):
            if col ==0 and row==0:
                continue

            axin1 = ax.inset_axes([col*width, 1-height*(row+1), width, height], facecolor='None')

            # hide xaxis and yaxis
            axin1.get_yaxis().set_visible(False)
            axin1.get_xaxis().set_visible(False)

            for edge, spine in axin1.spines.items():
                spine.set_visible(False)

            if row+col*len(table_min.index) < len(table_date_min.values):
                # plot the linechart
                # minimum value
                axin1.plot(min_date_value[col][row], color='blue', linewidth=1)
                # maximum value
                axin1.plot(max_date_value[col][row], color='green', linewidth=1)

                axes =fig.get_axes()
                axes[0]=axin
                axes[1]=axin1
                #set share x and y
                axes[0].get_shared_x_axes().join(axes[0], axes[1])
                axes[0].get_shared_y_axes().join(axes[0], axes[1])
                #axin.get_shared_y_axes().join(axin1, axin)
                axes[0].autoscale()
            else:
                break

# switch function
class Index(object):
    def __init__(self, ax, **kwargs):
        self.ax = ax
        self.kwargs = kwargs

    def display_min(self, event):
        self.ax.imshow(table_min.values, **self.kwargs)
        plt.draw()

    def display_max(self, event):
        self.ax.imshow(table_max.values, **self.kwargs)
        plt.draw()

# draw heatmap
im = heatmap(table_min.values, index, cols, ax=ax,
                   cmap=cmap, norm=norm)

# set hovertext here
hover = getHoverText()
cursorText = mp.cursor(im, hover=True)
@cursorText.connect("add")
def _(sel):
    sel.annotation.get_bbox_patch().set(fc='w', alpha=0.8)
    sel.annotation.set_text(hover[sel.target.index[0]][sel.target.index[1]])
    sel.annotation.set_zorder(20)

# draw linecharts
sublinechart(im)

# define button to switch
callback = Index(ax, cmap=cmap, norm=norm)
ax_min = plt.axes([0.35, 0.002, 0.1, 0.05])
ax_max = plt.axes([0.5, 0.002, 0.1, 0.05])
b_min = Button(ax_min, 'MinValue', color='#66C2A5', hovercolor='#3288BD')
b_min.on_clicked(callback.display_min)
b_max = Button(ax_max, 'MaxValue', color='#D53E4F', hovercolor='#9E0142')
b_max.on_clicked(callback.display_max)

#plt.tight_layout()
plt.show()

```
